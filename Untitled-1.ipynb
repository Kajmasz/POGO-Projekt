```python
# Function to extract Pokémon settings
def extract_pokemon_settings(data):
    pokemon_settings = data.get("data", {}).get("pokemonSettings", {})

    # Lista przechowująca standardowe oraz Mega Pokemony
    pokemon_entries = []
    
    # Tworzymy standardowego Pokémona (bez Mega Ewolucji)
    base_pokemon = {
        "pokemonTemplateId": data.get("templateId"),
        "pokemonPokedexNr": int(str(data.get("templateId"))[1:5]),
        "pokemonId":f"#{(str(data.get("templateId")))[1:5]}",
        "pokemonFamily": pokemon_settings.get("familyId"),
        "pokemonName": f"{(str(data.get("templateId")))[14:]}",
        "pokemonType": pokemon_settings.get("type"),
        "pokemonType2": pokemon_settings.get("type2"),
        "pokemonBaseAttack": pokemon_settings.get("stats", {}).get("baseAttack"),
        "pokemonBaseDefense": pokemon_settings.get("stats", {}).get("baseDefense"),
        "pokemonBaseStamina": pokemon_settings.get("stats", {}).get("baseStamina"),
        "pokemonHeightM": pokemon_settings.get("pokedexHeightM"),
        "pokemonWeightKg": pokemon_settings.get("pokedexWeightKg"),
        "pokemonCandyToEvolve": pokemon_settings.get("candyToEvolve"),
        "pokemonKmBuddyDistance": pokemon_settings.get("kmBuddyDistance"),
        "pokemonThirdMoveStardustToUnlock": pokemon_settings.get("thirdMove", {}).get("stardustToUnlock"),
        "pokemonThirdMoveCandyToUnlock": pokemon_settings.get("thirdMove", {}).get("candyToUnlock"),
        "pokemonShadowPurificationStardustNeeded": pokemon_settings.get("shadow", {}).get("purificationStardustNeeded"),
        "pokemonShadowPurificationCandyNeeded": pokemon_settings.get("shadow", {}).get("purificationCandyNeeded"),
        "pokemonQuickMoves": pokemon_settings.get("quickMoves"),
        "pokemonCinematicMoves": pokemon_settings.get("cinematicMoves"),
        "pokemonEliteQuickMove": pokemon_settings.get("eliteQuickMove"),
        "pokemonEliteCinematicMove": pokemon_settings.get("eliteCinematicMove"),
        "isMegaEvolution": False  # Oznaczamy, że to normalna forma
    }

    # Dodajemy podstawowego Pokémona do listy
    pokemon_entries.append(base_pokemon)

    # Tworzymy osobne wpisy dla każdej Mega Ewolucji
    temp_evo_data = pokemon_settings.get("tempEvoOverrides", [])
    for evo in temp_evo_data:
        mega_pokemon = {
            "pokemonTemplateId": f"{base_pokemon['pokemonTemplateId']}_MEGA",
            "pokemonId":f"#{(str(data.get("templateId")))[1:5]}{'M' if len((str(evo.get('tempEvoId')))[15:]) <= 4 else (str(evo.get('tempEvoId')))[20:]}",
            "pokemonPokedexNr": int(str(data.get("templateId"))[1:5]),
            "pokemonFamily": base_pokemon["pokemonFamily"],
            "pokemonName": f"{(str(data.get("templateId")))[14:]} {(str(evo.get('tempEvoId')))[15:]}",
            "pokemonType": evo.get("typeOverride1", base_pokemon["pokemonType"]),
            "pokemonType2": evo.get("typeOverride2", base_pokemon["pokemonType2"]),
            "pokemonBaseAttack": evo.get("stats", {}).get("baseAttack"),
            "pokemonBaseDefense": evo.get("stats", {}).get("baseDefense"),
            "pokemonBaseStamina": evo.get("stats", {}).get("baseStamina"),
            "pokemonHeightM": evo.get("averageHeightM"),
            "pokemonWeightKg": evo.get("averageWeightKg"),
            "pokemonCandyToEvolve": None,  # Mega Ewolucje nie ewoluują dalej
            "pokemonKmBuddyDistance": base_pokemon["pokemonKmBuddyDistance"],
            "pokemonThirdMoveStardustToUnlock": None,  
            "pokemonThirdMoveCandyToUnlock": None,
            "pokemonShadowPurificationStardustNeeded": None,
            "pokemonShadowPurificationCandyNeeded": None,
            "pokemonQuickMoves": base_pokemon["pokemonQuickMoves"],  # Mega formy używają tych samych ataków
            "pokemonCinematicMoves": base_pokemon["pokemonCinematicMoves"],
            "pokemonEliteQuickMove": base_pokemon["pokemonEliteQuickMove"],
            "pokemonEliteCinematicMove": base_pokemon["pokemonEliteCinematicMove"],
            "isMegaEvolution": True  # Oznaczamy, że to Mega forma
        }

        # Dodajemy Mega Pokémona do listy
        pokemon_entries.append(mega_pokemon)

    return pokemon_entries

# Function to extract move settings
def extract_move_settings(data):
    move_settings = data.get("data", {}).get("moveSettings", {})
    return {
        "moveId": move_settings.get("movementId"),
        "moveType": move_settings.get("pokemonType"),
        "movePower": move_settings.get("power"),
        "moveDurationMs": move_settings.get("durationMs"),
        "moveDamageWindowStartMs": move_settings.get("damageWindowStartMs"),
        "moveDamageWindowEndMs": move_settings.get("damageWindowEndMs"),
        "moveEnergy": move_settings.get("energyDelta"),
    }

# Function to extract and save moves data to a file
```

```python
all_moves = []

# Loop over each Pokémon data entry to extract move settings
for entry in data:
    move_data = extract_move_settings(entry)
    all_moves.append(move_data)

# Save the moves data to a separate JSON file
with open("expanded_pokemon_moves.json", "w", encoding="utf-8") as file:
    json.dump(all_moves, file, ensure_ascii=False, indent=4)
```

```python
all_pokemon = []
for entry in data:
    all_pokemon.extend(extract_pokemon_settings(entry))  # Dodajemy wszystkie formy

# Zapisujemy całą bazę z Mega Ewolucjami jako oddzielne wpisy
with open("expanded_pokemon_db.json", "w", encoding="utf-8") as file:
    json.dump(all_pokemon, file, indent=4)
```
