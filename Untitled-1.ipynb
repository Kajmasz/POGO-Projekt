```python
# Function to extract Pokémon settings
def extract_pokemon_settings(data):
    pokemon_settings = data.get("data", {}).get("pokemonSettings", {})
   
    return {
        "templateId": data.get("templateId"),
        "pokemonId": pokemon_settings.get("pokemonId"),
        "type": pokemon_settings.get("type"),
        "type2": pokemon_settings.get("type2"),
        "baseAttack": pokemon_settings.get("stats", {}).get("baseAttack"),
        "baseDefense": pokemon_settings.get("stats", {}).get("baseDefense"),
        "baseStamina": pokemon_settings.get("stats", {}).get("baseStamina"),
        "pokedexHeightM": pokemon_settings.get("pokedexHeightM"),
        "pokedexWeightKg": pokemon_settings.get("pokedexWeightKg"),
        "candyToEvolve": pokemon_settings.get("candyToEvolve"),
        "kmBuddyDistance": pokemon_settings.get("kmBuddyDistance"),
        "thirdMoveStardustToUnlock": pokemon_settings.get("thirdMove", {}).get("stardustToUnlock"),
        "thirdMoveCandyToUnlock": pokemon_settings.get("thirdMove", {}).get("candyToUnlock"),
        "shadowPurificationStardustNeeded": pokemon_settings.get("shadow", {}).get("purificationStardustNeeded"),
        "shadowPurificationCandyNeeded": pokemon_settings.get("shadow", {}).get("purificationCandyNeeded"),
        "quickMoves": pokemon_settings.get("quickMoves"),
        "cinematicMoves": pokemon_settings.get("cinematicMoves"),
        "eliteQuickMove": pokemon_settings.get("eliteQuickMove"),
        "eliteCinematicMove": pokemon_settings.get("eliteCinematicMove"),
        "megaId": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("tempEvoId"),
        "megaAttack": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("stats", {}).get("baseAttack"),
        "megaDefense": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("stats", {}).get("baseDefense"),
        "megaStamina": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("stats", {}).get("baseStamina"),
        "megaHeightM": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("averageHeightM"),
        "megaWeightKg": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("averageWeightKg"),
        "megaType1": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("typeOverride1"),
        "megaType2": pokemon_settings.get("tempEvoOverrides", [{}])[0].get("typeOverride2")
    }

# Function to extract move settings
def extract_move_settings(data):
    move_settings = data.get("data", {}).get("moveSettings", {})
    return {
        "movementId": move_settings.get("movementId"),
        "pokemonType": move_settings.get("pokemonType"),
        "power": move_settings.get("power"),
        "durationMs": move_settings.get("durationMs"),
        "damageWindowStartMs": move_settings.get("damageWindowStartMs"),
        "damageWindowEndMs": move_settings.get("damageWindowEndMs"),
        "energyDelta": move_settings.get("energyDelta"),
    }
```

```python
# Extract data from each JSON record
pokemon_data = []
move_data = []

for record in data:
    pokemon_data.append(extract_pokemon_settings(record))
    move_data.append(extract_move_settings(record))
```

```python
pokemon_df = pd.DataFrame(pokemon_data)
move_df = pd.DataFrame(move_data)

# Save to CSV or Excel
pokemon_df.to_csv('pokemon_data.csv', index=False)
move_df.to_csv('move_data.csv', index=False)

# Display DataFrames
pokemon_df
```

```python
# Process the JSON data and split it
pokemon_data = []
move_data = []

for record in data:
    pokemon = extract_pokemon_settings(record)
    move = extract_move_settings(record)
    
    # Only add non-empty dictionaries to respective lists
    if pokemon and any(pokemon.values()):
        pokemon_data.append(pokemon)
    if move and any(move.values()):
        move_data.append(move)

# Save Pokémon and moves data to separate JSON files
with open('pokemon_data.json', 'w') as pokemon_file:
    json.dump(pokemon_data, pokemon_file, indent=4)

with open('move_data.json', 'w') as move_file:
    json.dump(move_data, move_file, indent=4)

print("Pokémon data saved to 'pokemon_data.json'")
print("Move data saved to 'move_data.json'")
```
